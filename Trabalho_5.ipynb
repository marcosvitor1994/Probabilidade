```
# 5 - Escreva em Python um algoritmo que solicite para o usuário inserir
# notas de alunos entre 0 e 10 em um loop infinito com condição de parada
# uma nota igual a -999. A cada interação valide que a nota seja entre 
# 0 e 10 e solicite uma nota correta em caso de erro. 

# Após obter todas as notas, o programa deve mostrar. 

# A - Média aritmética simples.
# B - Média harmônic.
# C - Média geométrica.
# D - Moda.
# E - Variância.
# F - Desvio padrão.
# G - Coeficiente de variância.

import statistics
from scipy.stats.mstats import gmean

lista_4 = [] #Lista com a podrução em kg. 


print('Digite a(s) nota(s) do aluno \nPara Sair digite[-999]') #Menu do programa
while True:#Loop infinito 

  x1 = float(input("Digite a(s) nota(s): "))#Variavel que receberá as notas
  if x1 == -999:
    break
  if x1 >= 0 and x1 <= 10:
    lista_4.append(x1)#Adicionando as notas a lista
  if x1 < 0 or x1 > 10:
    print('Erro, Digite uma nota válida.')
    x1 = int(input("Digite a(s) nota(s): "))


print(f'A - Média aritmética simples: {statistics.mean(lista_4)}') #Média aritmética utilizando a biblioteca stastistics 
print(f'B - A Média harmônica é: {statistics.harmonic_mean(lista_4)}') #Média harmônica utilizando a biblioteca stastistics, se o valor de alguma nota for 0 dará erro na conta
print(f'C - A média geometríca é: {gmean(lista_4)}') #Calculo feito utilizando a biblioteca scipy.stats
print(f'D - A moda é: {statistics.mode(lista_4)}') #Moda utilizando a biblioteca stastistics e depende das notas que forem colocar pode dar erro ou não. 8
print(f'E - A variância é: {statistics.variance(lista_4)}') #Variância utilizando a biblioteca stastistics
print(f'F - O desvio padrão é: {statistics.stdev(lista_4)}') #Desvio padrão utilizando a biblioteca stastistics
```
